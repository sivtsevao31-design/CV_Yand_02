```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO
import torch
```

```python
device = 0 if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")
```

```python
def train_model():
    model = YOLO('yolov8s-seg.pt')

    model.train(
        data='C:/Users/bogda/AiLessons1/dataset/data.yaml',
        epochs=150,
        imgsz=320,
        batch=16,
        device=device,
        workers=4,
        
        mosaic=1.0,
        mixup=0.1,
        fliplr=0.5,
        flipud=0.0,
        degrees=15.0,
        translate=0.1,
        scale=0.5,
        shear=2.0,
        hsv_h=0.015,
        hsv_s=0.7,
        hsv_v=0.4,
        
        patience=30,
        project='animal_seg',
        name='rtx3050_v1'
    )

if __name__ == '__main__':
    train_model()
```

```python
model = YOLO('runs/segment/animal_seg/rtx3050_v1-2/weights/best.pt') 
metrics = model.val(split='test') 
print(f"Точность боксов (mAP50): {metrics.box.map50}")
print(f"Точность масок (mAP50): {metrics.seg.map50}")
```
